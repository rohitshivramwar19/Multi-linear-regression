MODEL BUILDING


```python
x=df1.drop(["Profit"],axis=1).values
y=df1.iloc[:,3].values.reshape(-1,1)
x
```

```python
model=linear_model.LinearRegression()
model.fit(x,y)
```

```python
model.rank_
```

```python
model.coef_
```

```python
model.intercept_
```

```python
r2_score(y,model.predict(x))
```

PREDICTION 

```python
df2=pd.DataFrame({"R&D Spend":153450,"Administration":134300,"Marketing Spend":472000},index=[1])
model.predict(df2)
print("The profit will be:",model.predict(df2))
```

```python
df3=df1.rename({"Marketing Spend":"Marketing_Spend"},axis=1)
df4=df3.rename({"R&D Spend":"RnD_Spend"},axis=1)
df4.head()
```

```python
module=smf.ols("Profit~RnD_Spend+Administration+Marketing_Spend",data=df4).fit()
module.summary()
```

TEST FOR NORMALITY OF RESIDULES(Q-Q PLOT)

```python
import statsmodels.api as smf
qqplot=smf.qqplot(module.resid,line="q")
```

RESIDUAL PLOT FOR HOMOSCEDASTICITY

```python
def std(values):
    return (values - values.mean())/values.std()

plt.scatter(std(module.fittedvalues),
           std(module.resid))
```

RESIDUAL VS REGRESSOR

```python
fig=plt.figure(figsize=(15,8))
fig=smf.graphics.plot_regress_exog(module,"RnD_Spend",fig=fig)
plt.show()
```

```python
fig = plt.figure(figsize=(15,8))
fig = smf.graphics.plot_regress_exog(module, "Marketing_Spend", fig=fig)
plt.show()
```

```python
fig = plt.figure(figsize=(15,8))
fig = smf.graphics.plot_regress_exog(module, "Administration", fig=fig)
plt.show()
```

MODEL DELETION DIAGNOSTIC

DETACTING OUTLIERS OR INFLUNCERS 

```python
module_influence=module.get_influence()
(c, _)=module_influence.cooks_distance
```

```python
influence_plot(module)
plt.show()
```

```python
np.argmax(c),np.max(c)
```

Since the value is <1,we don't need to do the diagnostic process and finalize the model.
